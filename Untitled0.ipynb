```python
# Read the csv file and save it to a dataframe called "airlines_df"
airlines_df = pd.read_csv('airline_passenger_satisfaction.csv')
```

```python
#There are 393 rows with null values. 393/129,880 = 0.003 or .30%. Considering materiality, we will drop these rows. Now we have 129,487 instances. 
airlines_df = airlines_df.dropna()
airlines_df
```

```python
#Examine the dataset and Check for any special characters, as well as the range of unique numerical and categorical values in the dataset
for i in airlines_df:
  print('\n', i, '\n', airlines_df[i].unique())
```

```python
#Countplot to show the age range of passengers and the count of the passengers per age group

plt.figure(figsize = (20, 20))
plot = sns.countplot(x = "Age", data = airlines_df, dodge = True)
plot.set_title("Age of Passengers", weight = "bold", fontsize = 20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python

# DROP "ID" Field
airlines_df = airlines_df.drop(['ID'], axis = 1)
```

```python
#Convert object type to numerical labels: 
#Gender: Female(0), Male(1)
#Customer Type: First-time(0), Returning(1)
#Type of Travel: Business(0), Personal(1)
#Class: Business(0), Economy (1), Economy Plus (2)
#Satisfaction: Neutral or Dissatisfied(0), Satisfied(1)

airlines_df.head(20)
for i in airlines_df.columns:
  if airlines_df[i].dtype == "object":
    label_encoder = preprocessing.LabelEncoder()
    airlines_df[i] = label_encoder.fit_transform(airlines_df[i])

airlines_df
```

```python
# Select only numerical data to prepare for Correlation Analysis
airlines_numerics_df = airlines_df.select_dtypes(include = "number")
airlines_numerics_df
```

```python
# Plot correlation matrix between pairs of attributes to determine correlation
corr_matrix = airlines_numerics_df.corr()
corr_matrix
```

```python
# Plot correlation heatmap
mask = np.triu(np.ones_like(corr_matrix))
plt.figure(figsize = (15, 15))
num_heatmap = sns.heatmap(data = corr_matrix, mask = mask, cmap = "RdBu", vmax = 1.0, vmin = -1.0, linewidths = 0.1, square = True, cbar_kws={'shrink': 0.82})
plt.title('Correlation Heatmap: Numeric Features', fontweight = "bold", fontdict = {'fontsize': 16})
plt.show()
```
